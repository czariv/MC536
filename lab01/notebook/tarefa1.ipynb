```python
import io
import requests
import json
import xml.etree.ElementTree as et
```

##Primeira Fonte

Primeiramente vou importar uma api do pubchem em formato xml, montei o link da api de forma a obter alguns dados interessantes sobre alguns compostos químicos encontrados em algumas drogas ilicitas.

```python
url = "https://pubchem.ncbi.nlm.nih.gov/rest/pug/compound/cid/702,2543,446220,5462328/property/Title,MolecularFormula,MolecularWeight,CanonicalSMILES,IUPACName/xml"
data = requests.get(url).content
x = et.parse(io.StringIO(data.decode("utf-8")))
print(data.decode("utf-8"))
```

Depois analisei o resultado do request, imprimindo todos os dados para fazer uma analise. Decidi então encontrar o nome de cada composto especifica

```python
root=x.getroot()
print(root.tag)
```

Achando o nome da root endendi um pouco mais sobre a nomeclatura nessa api

```python
for child in root:
    print(child.tag)
    for c in child:
        print("- "+c.tag)
```

Depois eu achei o nome do caminho que eu teria que percorrer

```python
for child in root:
    print(child.find("{http://pubchem.ncbi.nlm.nih.gov/pug_rest}Title").text)
```

##Segunda Fonte

Agora vou fazer a analise dos dados obtidos na api da world health organization, estou pegando uma base de dados sobre uso de drogas em formato json.

```python
url = "https://apps.who.int/gho/athena/api/RSUDSUBSTANCEDEPENDENCE?filter=CONTRY:USA&format=json"
js = requests.get(url).content
print(js.decode("utf-8"))
```

Primeiro analisei o resultado do request, como feito anteriormente e decidi imprimir os tipos de vicios

```python
data = requests.get(url)
js=data.json()
for i in range(4):
    print(js["dimension"][0]["code"][i]["display"])
```
