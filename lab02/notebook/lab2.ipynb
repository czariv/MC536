## Tarefa 1

Carregue o nó de Tenda dos Milagres (da versão em inglês Tent of Miracles) em Python e mostre o conteúdo JSON.

```python
import io
import requests
import json
```

```python
url_tent = "https://dbpedia.org/data/Tent_of_Miracles_(novel).jsod"
data_tent = requests.get(url_tent)
json_tent = data_tent.json()
print(json.dumps(json_tent, indent=3))
```

## Tarefa 2

Recupere via Python e mostre o autor de Tenda dos Milagres

```python
autor_tent = json_tent["d"]["results"][0]["http://dbpedia.org/ontology/author"]["__deferred"]["uri"]
print("Autor: "+autor_tent)
```

## Tarefa 3

Quais os termos usados para Aspirina (consulta JSON/MeSH em Python)?

```python
url_aspirin = "https://id.nlm.nih.gov/mesh/lookup/descriptor?label=Aspirin&match=exact"
data_aspirin = requests.get(url_aspirin)
json_aspirin = data_aspirin.json()
print(json.dumps(json_aspirin, indent=3))
```

```python
uri_mesh = json_aspirin[0]["resource"]
bar = str.rindex(uri_mesh, "/")
code_mesh = uri_mesh[bar+1:]
url_template = "https://id.nlm.nih.gov/mesh/lookup/details?descriptor={code}"
url_mesh = url_template.replace("{code}", code_mesh)
data_mesh = requests.get(url_mesh)
json_mesh = data_mesh.json()
print(json.dumps(json_mesh, indent=3))
```

```python
for i in json_mesh["terms"]:
    print(i["label"])
```

## Tarefa 4

Escolha algum lugar no OpenStreetMap e elabore a mesma sequência de três chamadas de serviços para o seu conceito.

```python
url_colosseum = "http://nominatim.openstreetmap.org/search?format=json&q=colosseo"
data_colosseum = requests.get(url_colosseum)
json_colosseum = data_colosseum.json()
print(json.dumps(json_colosseum, indent=3))
```

```python
code_maps = str(json_colosseum[0]["osm_id"])
type_maps = json_colosseum[0]["osm_type"]
url_template = "https://api.openstreetmap.org/api/0.6/{element_type}/{id}.json"
url_maps = url_template.replace("{element_type}", type_maps)\
                       .replace("{id}", code_maps)
data_maps = requests.get(url_maps)
json_maps = data_maps.json()
print(json.dumps(json_maps, indent=3))
```

```python
code_wiki = json_maps["elements"][0]["tags"]["name:en"].replace(" ", "_")\
                                                         .replace("it:", "")
url_template = "http://dbpedia.org/data/{concept}.jsod"
url_wiki = url_template.replace("{concept}", code_wiki)
data_wiki = requests.get(url_wiki)
json_wiki = data_wiki.json()
print(json.dumps(json_wiki, indent=3))
```
